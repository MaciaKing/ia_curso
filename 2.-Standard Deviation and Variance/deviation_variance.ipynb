# Standard Deviation and Variance

```python
import numpy as np
import matplotlib.pyplot as plt

incomes = np.random.normal(100, 20, 10000)
plt.hist(incomes, 50)
plt.show()
```

```python
# Standard deviation
incomes.std()
```

```python
# Variance
incomes.var()
```
